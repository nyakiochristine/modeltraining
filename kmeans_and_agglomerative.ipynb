```python
from sklearn.datasets import load_iris
iris_dataset=load_iris()
```

```python
X=iris_dataset.data
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
model =KMeans(n_clusters=3,init='k-means++',max_iter=300,n_init=10,random_state=0)
model.fit(X)
```

```python
print("Cluster memberships:",model.labels_)#predicted values
#cluster centers central point kmeans assumes cluster is circular in shape
print("Cluster Centers:",model.cluster_centers_)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=model.labels_,cmap='rainbow')
```

```python
plt.plot(model.cluster_centers_[0][0],model.cluster_centers_[0][1],'ro',markersize=16,alpha =0.5,label='')
plt.plot(model.cluster_centers_[1][0],model.cluster_centers_[1][1],'ro',markersize=16,alpha =0.5)
plt.plot(model.cluster_centers_[2][0],model.cluster_centers_[2][1],'ro',markersize=16,alpha =0.5)
```

```python
plt.show()
```

```python
from sklearn.datasets import load_iris
iris_dataset=load_iris()
```

```python
X=iris_dataset.data
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X)
```

```python
from sklearn.cluster import AgglomerativeClustering
model=AgglomerativeClustering()
model.fit(X)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=model.labels_,cmap='rainbow')

plt.show()
```

```python
#import the dendrogram and the ward clustering function from Scipi
from scipy.cluster.hierarchy import dendrogram,ward
# Implements Ward's method, which is a hierarchical clustering algorithm that minimizes the variance of the clusters being merged.
linkage_array=ward(X)

#plot dendrogram for linkage_array containing distances
dendrogram(linkage_array)
#mark the cuts in the tree
plt.axhline(y=8,c='k')
plt.axhline(y=12,c='k')#line 1
plt.xlabel('Sample Index')
#distance between merged clusters
plt.ylabel(' Cluster Distance')
```

```python
#dbscan
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris_dataset=load_iris()
X=iris_dataset.data
```

```python
sc=StandardScaler()
X=sc.fit_transform(X)
```

```python
from sklearn.cluster import AgglomerativeClustering
#esp-distance between samples and minimum samples
model=DBSCAN(eps=1.5,min_samples=3)
model.fit(X)
```

```python
assignment =model.fit_predict(X)
print("Cluster memberships", assignment)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=model.labels_,cmap='rainbow')

plt.show()
```
